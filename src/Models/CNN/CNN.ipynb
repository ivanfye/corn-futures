```python
import os

data_path = '/content/drive/MyDrive/U of T/CSC413H1 S/CSC413 Final Project'
os.chdir(f'{data_path}/Models/CNN')

from CNN import *
```

# Setup

```python
# Get cleaned data
inputs, labels, dates = get_cleaned_data(data_path)

# Divide inputs by type
inputs = {
    'US Related Crops': inputs[0:4],
    'US Weather': inputs[4:8],
    'International Corn and Weather': inputs[8:14],
}
```

# Training

```python
# Set dataset creation parameters
input_length = 18
offset = 0

# Mean elapsed time in days for the specified input_length
print(np.mean(dates[input_length:] - dates[:-input_length]) / np.timedelta64(1, 'D'))
```

```python
# Train all models
for predictors in inputs.keys():
    for i in range(10):
        print('--------------------------------------------------')
        print(f'{predictors} Prediction {offset} Days Ahead Model {i + 1}')

        # Create training, validation, and test sets
        train_data = create_dataset(inputs[predictors], labels, dates, np.datetime64('2014-01-01'), np.datetime64('2021-12-31'), input_length, offset, overlap=True)
        val_data = create_dataset(inputs[predictors], labels, dates, np.datetime64('2022-01-01'), np.datetime64('2022-12-31'), input_length, offset, overlap=True)
        test_data = create_dataset(inputs[predictors], labels, dates, np.datetime64('2023-01-01'), np.datetime64('2023-12-31'), input_length, offset, overlap=True)

        # Length of each set
        print(f'Train: {len(train_data)} Val: {len(val_data)} Test: {len(test_data)}')

        # Train model
        model = CNN(len(inputs[predictors]), input_length)
        model = train_model(model, train_data, val_data, batch_size=64, learning_rate=0.001, weight_decay=0.0001, num_epochs=100, plot_every=25)
        test_loss(model, test_data)
        print('--------------------------------------------------')
```
